```python
import os
import pandas as pd
from statsmodels.tsa.stattools import adfuller

# Paths to the data folders
data_paths = {
    'california_electricity_prices': '/Users/aashrithasankineni/Downloads/california_electricity_prices',
    'california_carbon_footprint': '/Users/aashrithasankineni/Downloads/california_carbon_footprint',
    'texas_electricity_prices': '/Users/aashrithasankineni/Downloads/texas_electricity_prices',
    'texas_carbon_footprint': '/Users/aashrithasankineni/Downloads/texas_carbon_footprint',
    'newyork_electricity_prices': '/Users/aashrithasankineni/Downloads/newyork_electricity_prices',
    'newyork_carbon_footprint': '/Users/aashrithasankineni/Downloads/newyork_carbon_footprint'
}

# Configuration for electricity files based on sample files inspected
electricity_config = {
    "california_electricity_prices": {"skip_rows": 3, "datetime_col": "UTC Timestamp (Interval Ending)", "price_col": "PALOVRDE_ASR-APND LMP"},
    "texas_electricity_prices": {"skip_rows": 3, "datetime_col": "UTC Timestamp (Interval Ending)", "price_col": "Hub average LMP"},
    "newyork_electricity_prices": {"skip_rows": 3, "datetime_col": "UTC Timestamp (Interval Ending)", "price_col": "A - West LMP"}
}

# Dictionary to store ARIMA-prepared dataframes
arima_ready_data = {}

# ARIMA-specific preprocessing function
def preprocess_for_arima(df, target_column):
    # Convert datetime column to datetime format
    df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
    df.set_index('datetime', inplace=True)
    
    # Ensure target column data is numeric and handle missing values
    df[target_column] = pd.to_numeric(df[target_column], errors='coerce')
    df[target_column] = df[target_column].interpolate(method='linear').ffill().bfill()
    
    # Check stationarity with ADF test; apply differencing if needed
    result = adfuller(df[target_column].dropna())
    if result[1] > 0.05:  # p-value > 0.05 suggests non-stationarity
        df[target_column] = df[target_column].diff().dropna()
    
    return df

# Process each dataset for ARIMA preprocessing
for folder_name, folder_path in data_paths.items():
    # Determine whether it's an electricity price or carbon footprint folder
    is_electricity = 'electricity' in folder_name
    target_column = 'marginal_price' if is_electricity else 'carbon_intensity'
    
    # Loop through each file in the folder
    for filename in os.listdir(folder_path):
        file_path = os.path.join(folder_path, filename)
        
        # Process only .csv files
        if filename.endswith('.csv'):
            try:
                # Handle electricity files with specific configurations
                if is_electricity:
                    config = electricity_config.get(folder_name, None)
                    if config:
                        # Load file, skipping metadata rows
                        df = pd.read_csv(file_path, skiprows=config["skip_rows"])
                        
                        # Check if required columns are present
                        if config["datetime_col"] in df.columns and config["price_col"] in df.columns:
                            df = df[[config["datetime_col"], config["price_col"]]]
                            df.columns = ['datetime', target_column]
                        else:
                            print(f"Skipping {filename} due to missing columns.")
                            continue
                    else:
                        print(f"No configuration for {folder_name}. Skipping file.")
                        continue

                # Handle carbon footprint files (assumes datetime in 1st column, carbon intensity in 6th column if no headers)
                else:
                    df = pd.read_csv(file_path, on_bad_lines='skip')
                    if 'datetime' in df.columns and 'carbon_intensity' in df.columns:
                        df = df[['datetime', 'carbon_intensity']]
                    elif df.shape[1] > 5:
                        df = df.iloc[:, [0, 5]]
                        df.columns = ['datetime', target_column]
                    else:
                        print(f"Skipping {filename} due to insufficient columns.")
                        continue
                
                # Preprocess data for ARIMA
                processed_df = preprocess_for_arima(df, target_column)
                if processed_df is not None:
                    arima_ready_data[file_path] = processed_df
                    print(f"ARIMA preprocessing completed for {file_path}")
            except Exception as e:
                print(f"Error processing {file_path}: {e}")

# Display processed data
for name, df in arima_ready_data.items():
    print(f"\nProcessed data for ARIMA in {name}:\n", df.head())
```

```python
import os
import pandas as pd

# Base folder paths
base_folders = {
    "california": "/Users/aashrithasankineni/Downloads/california_carbon_footprint",
    "newyork": "/Users/aashrithasankineni/Downloads/newyork_carbon_footprint",
    "texas": "/Users/aashrithasankineni/Downloads/texas_carbon_footprint",
}

# Helper function to load files from a folder
def load_files_from_folder(folder_path):
    return [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(".csv")]

# Helper function to load, clean, and resample data
def process_carbon_files(files, time_col="Datetime (UTC)", freq="H"):
    combined_df = []
    for file in files:
        try:
            print(f"Processing file: {file}")
            df = pd.read_csv(file)
            
            if df.empty:
                print(f"Warning: {file} is empty.")
                continue
            
            # Ensure the timestamp column is properly parsed
            if time_col not in df.columns:
                print(f"Error: {time_col} not found in {file}. Available columns: {df.columns}")
                continue
            df[time_col] = pd.to_datetime(df[time_col], errors="coerce")
            df.set_index(time_col, inplace=True)
            
            # Filter numeric columns only
            numeric_df = df.select_dtypes(include=["number"])
            if numeric_df.empty:
                print(f"Warning: No numeric columns found in {file}. Skipping.")
                continue
            
            # Resample to hourly frequency
            numeric_df = numeric_df.resample(freq).mean()
            combined_df.append(numeric_df)
        
        except Exception as e:
            print(f"Error processing file {file}: {e}")
    
    if not combined_df:
        raise ValueError("No valid data found in the provided files.")
    
    return pd.concat(combined_df)

# Load and process carbon footprint data for California, New York, and Texas
try:
    california_files = load_files_from_folder(base_folders["california"])
    california_carbon = process_carbon_files(california_files)
    print("California carbon data processed successfully.")
except ValueError as ve:
    print(f"Critical Error (California): {ve}")

try:
    newyork_files = load_files_from_folder(base_folders["newyork"])
    newyork_carbon = process_carbon_files(newyork_files)
    print("New York carbon data processed successfully.")
except ValueError as ve:
    print(f"Critical Error (New York): {ve}")

try:
    texas_files = load_files_from_folder(base_folders["texas"])
    texas_carbon = process_carbon_files(texas_files)
    print("Texas carbon data processed successfully.")
except ValueError as ve:
    print(f"Critical Error (Texas): {ve}")
```

```python
import pandas as pd
import os

def detect_header(file_path, keyword="UTC Timestamp"):
    """
    Detect the header row dynamically by looking for a keyword in the file.
    """
    with open(file_path, 'r') as file:
        for i, line in enumerate(file):
            if keyword in line:
                return i
    return None  # Return None if the keyword is not found

def process_price_files_dynamic(files, time_col="UTC Timestamp (Interval Ending)", freq="H"):
    combined_df = []
    for file in files:
        try:
            print(f"Processing file: {file}")
            
            # Detect the header row dynamically
            header_row = detect_header(file, keyword="UTC Timestamp")
            if header_row is None:
                print(f"Error: Unable to detect header in {file}. Skipping.")
                continue
            
            # Read the file with the detected header
            df = pd.read_csv(file, skiprows=header_row)
            
            # Ensure the timestamp column exists
            if time_col not in df.columns:
                print(f"Error: {time_col} not found in {file}. Columns: {df.columns}")
                continue

            # Parse timestamps and set index
            df[time_col] = pd.to_datetime(df[time_col], errors="coerce")
            df.set_index(time_col, inplace=True)
            
            # Select LMP columns (columns containing "LMP")
            lmp_columns = [col for col in df.columns if "LMP" in col]
            if not lmp_columns:
                print(f"Warning: No LMP columns found in {file}. Skipping.")
                continue

            # Aggregate LMP columns (mean of all LMP columns)
            df["Average LMP"] = df[lmp_columns].mean(axis=1)
            
            # Resample to hourly frequency
            hourly_df = df["Average LMP"].resample(freq).mean()
            combined_df.append(hourly_df)
        
        except Exception as e:
            print(f"Error processing file {file}: {e}")
    
    if not combined_df:
        raise ValueError("No valid data found in the provided files.")
    
    return pd.concat(combined_df)

# Define paths to electricity price folders
price_folders = {
    "california": "/Users/aashrithasankineni/Downloads/california_electricity_prices",
    "newyork": "/Users/aashrithasankineni/Downloads/newyork_electricity_prices",
    "texas": "/Users/aashrithasankineni/Downloads/texas_electricity_prices",
}

# Load files from folders
def load_files_from_folder(folder_path):
    return [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(".csv")]

# Process files for each region
try:
    california_files = load_files_from_folder(price_folders["california"])
    california_prices = process_price_files_dynamic(california_files)
    print("California electricity price data processed successfully.")
except ValueError as ve:
    print(f"Critical Error (California Prices): {ve}")

try:
    newyork_files = load_files_from_folder(price_folders["newyork"])
    newyork_prices = process_price_files_dynamic(newyork_files)
    print("New York electricity price data processed successfully.")
except ValueError as ve:
    print(f"Critical Error (New York Prices): {ve}")

try:
    texas_files = load_files_from_folder(price_folders["texas"])
    texas_prices = process_price_files_dynamic(texas_files)
    print("Texas electricity price data processed successfully.")
except ValueError as ve:
    print(f"Critical Error (Texas Prices): {ve}")
```

```python
import pandas as pd
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# Load datasets (update file paths as needed)
data_2021 = pd.read_csv('/Users/aashrithasankineni/Downloads/california_carbon_footprint/US-CAL-CISO_2021_hourly.csv')
data_2022 = pd.read_csv('/Users/aashrithasankineni/Downloads/california_carbon_footprint/US-CAL-CISO_2022_hourly.csv')
data_2023 = pd.read_csv('/Users/aashrithasankineni/Downloads/california_carbon_footprint/US-CAL-CISO_2023_hourly.csv')

# Combine datasets
data_combined = pd.concat([data_2021, data_2022, data_2023], ignore_index=True)

# Convert Datetime to pandas datetime and set as index
data_combined['Datetime (UTC)'] = pd.to_datetime(data_combined['Datetime (UTC)'])
data_combined.set_index('Datetime (UTC)', inplace=True)

# Explicitly set frequency for time index (hourly)
data_combined = data_combined.asfreq('h')

# Select the target column
carbon_intensity = data_combined['Carbon Intensity gCO₂eq/kWh (direct)']

# Plot ACF and PACF for non-seasonal components
plt.figure(figsize=(12, 6))
plt.subplot(121)
plot_acf(carbon_intensity.dropna(), lags=48, ax=plt.gca(), title="ACF (Autocorrelation)")
plt.subplot(122)
plot_pacf(carbon_intensity.dropna(), lags=48, ax=plt.gca(), title="PACF (Partial Autocorrelation)")
plt.tight_layout()
plt.show()

# Fit a SARIMA model with seasonal components
# Seasonal order (P, D, Q, S): Choose S=24 for hourly seasonality
print("Fitting SARIMA model...")
sarima_model = SARIMAX(
    carbon_intensity,
    order=(1, 1, 1),  # Non-seasonal (p, d, q)
    seasonal_order=(1, 1, 1, 24),  # Seasonal (P, D, Q, S)
    enforce_stationarity=False,
    enforce_invertibility=False
)
sarima_fit = sarima_model.fit(disp=False)

# Display model summary
print(sarima_fit.summary())

# Forecast for 2024 (8760 hours for a leap year)
forecast_steps = 8760  # Number of hours in 2024
future_dates = pd.date_range(start='2024-01-01 00:00:00', periods=forecast_steps, freq='h')
forecast_2024 = sarima_fit.forecast(steps=forecast_steps)

# Create a DataFrame for forecasted results
forecast_df = pd.DataFrame({'Datetime (UTC)': future_dates, 'Forecast': forecast_2024})
forecast_df.set_index('Datetime (UTC)', inplace=True)

# Save forecast to a CSV file
forecast_df.to_csv('/Users/aashrithasankineni/Downloads/forecast_2024_hourly_sarima.csv')

# Display a preview of the forecast
print("Hourly Forecast Preview (2024):")
print(forecast_df.head())
```
